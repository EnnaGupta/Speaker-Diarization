Audio Deserialisation

```python
# Import the AudioSegment class from pydub for audio file manipulation.
from pydub import AudioSegment
import io
# Function to deserialize the audio file to an AudioSegment object.
def deserialize_audio(file_path):
    # Load the audio file into an AudioSegment object
    audio = AudioSegment.from_file(file_path)

   # Print details about the audio file for verification.
    print(f"Channels: {audio.channels}")
    print(f"Channels: {audio.channels}")
    print(f"Frame rate: {audio.frame_rate}")
    print(f"Sample width: {audio.sample_width}")
    print(f"Duration (ms): {len(audio)}")

    return audio

# Load the audio file into an AudioSegment object.
audio_file_path = "/content/WAudio_1.wav"  # replace with your audio file path
audio = deserialize_audio(audio_file_path)

# Now, 'audio' is an AudioSegment object that you can manipulate or play back
```

Transcription:The Whisper model (large-v3) is loaded with GPU support for transcription.
Transcription is performed on the audio file, and the time taken is measured.
Transcription segments are extracted from the result.


```python
# Load Whisper model with GPU support for transcription.
import whisper
model = whisper.load_model("large-v3", device=device)
audio_file = "/content/WAudio_1.wav"
# Perform transcription with timing on the audio file and measure the time taken.
start_time = time.time()
print(f"Transcription took {time.time() - start_time:.2f} seconds")
result = model.transcribe(audio_file)
# Extract transcriptions from the Whisper model result.
segments = result["segments"]
```

```python
# Mapping Transcriptions to Speakers:A set processed_segments is initialized to keep track of processed transcription segments to avoid duplication.
#Iterates through diarization results and maps each transcription segment to the corresponding speaker based on time overlap.
#Transcription texts are joined together for each speaker and printed.

# Initialize a set to keep track of processed transcription segments to avoid duplication.
processed_segments = set()

# Iterate through diarization results and map each transcription segment to the corresponding speaker.
for segment, _, speaker in diarization.itertracks(yield_label=True):
    start_time = segment.start
    end_time = segment.end

    # Find the transcription segments that overlap with the diarization segment.
    transcription_texts = []
    for seg in segments:
        if seg['start'] < end_time and seg['end'] > start_time and seg['text'] not in processed_segments:
            transcription_texts.append(seg['text'])
            processed_segments.add(seg['text'])  # Add processed segment to set

    # If there are transcription texts for this segment
    if transcription_texts:
        # Join the texts together into a single string.
        combined_text = " ".join(transcription_texts)
        print(f'Speaker "{speaker}" - "{combined_text}"')
```

```python
import subprocess

def get_version(package_name):
    result = subprocess.run(['pip', 'show', package_name], stdout=subprocess.PIPE)
    for line in result.stdout.decode().split('\n'):
        if line.startswith('Version:'):
            return line.split(' ')[1]


print("pyannote.audio version:", get_version('pyannote.audio'))
print("pydub version:", get_version('pydub'))

```
